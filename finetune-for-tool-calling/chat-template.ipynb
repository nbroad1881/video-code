```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen3-14B")
```

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "What is the capital of Canada?"},
    {"role": "assistant", "content": "The capital of Canada is Ottawa."},
]

print(tokenizer.apply_chat_template(messages, tokenize=False))
```

```python
messages = [
    {"role": "system", "content": "You area helpful assistant."},
    {"role": "user", "content": "What is the capital of Canada?"},
    {
        "role": "assistant",
        "reasoning_content": "I need to respond with the capital of Canada. I know it's Ottawa. I'll respond now.",
        "content": "The capital of Canada is Ottawa.",
    },
]
print(tokenizer.apply_chat_template(messages, tokenize=False))
```

```python
messages = [
    {"role": "system", "content": "You area helpful assistant."},
    {"role": "user", "content": "What is the capital of Canada?"},
    {
        "role": "assistant",
        "reasoning_content": "I need to respond with the capital of Canada. I know it's Ottawa. I'll respond now.",
        "content": "The capital of Canada is Ottawa.",
    },
    {"role": "user", "content": "What keeps the moon in orbit around the earth?"},
    {
        "role": "assistant",
        "reasoning_content": "The user is asking about the moon. I need to respond with the correct answer. I know it's gravity. I'll respond now.",
        "content": "The moon is kept in orbit around the earth by gravity.",
    },
]
print(tokenizer.apply_chat_template(messages, tokenize=False))
```

```python
tools = [
    {
        "type": "function",
        "function": {
            "name": "get_crypto_price",
            "description": "Get the current price of a cryptocurrency",
            "parameters": {
                "type": "object",
                "properties": {
                    "symbol": {
                        "type": "string",
                        "description": "The ticker symbol, e.g. BTC, ETH",
                    }
                },
                "required": ["symbol"],
            },
        },
    }
]

messages = [
    {"role": "user", "content": "What is the current price of Bitcoin?"},
    {
        "role": "assistant",
        "reasoning_content": "I need to get the current price of Bitcoin. I should use the get_crypto_price tool.",
        "tool_calls": [
            {
                "id": "call_12345xyz",
                "type": "function",
                "function": {
                    "name": "get_crypto_price",
                    "arguments": '{"symbol": "BTC"}',
                },
            }
        ],
    },
    {"role": "tool", "tool_call_id": "call_12345xyz", "content": '{"price": 10000}'},
    {"role": "assistant", "content": "The current price of Bitcoin is $10,000."},
]

print(tokenizer.apply_chat_template(messages, tokenize=False, tools=tools))
```
